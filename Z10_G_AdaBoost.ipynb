```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from mlxtend.plotting import plot_decision_regions
```

# AdaBoost

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=500, noise=0.30, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

ada_clf = AdaBoostClassifier(
    DecisionTreeClassifier(max_depth=1), 
    n_estimators=1, learning_rate=0.5, 
    algorithm="SAMME.R", random_state=42)
ada_clf.fit(X_train, y_train)
```

```python
plot_decision_regions(X, y, ada_clf)
plt.show()
```

# Zad.

Sprawdż różne parametry dla AdaBoostClassifier

* n_estimators=2, learning_rate=0.5
* n_estimators=2, learning_rate=1
* n_estimators=10, learning_rate=0.5
* n_estimators=10, learning_rate=1

```python
params_lst = [
  {
      "n_estimators": 2,
      "learning_rate": 0.5
  },
  {
      "n_estimators": 2,
      "learning_rate": 1
  },
  {
      "n_estimators": 10,
      "learning_rate": 0.5
  },
  {
      "n_estimators": 10,
      "learning_rate": 1
  }
]

for params in params_lst:
  print(params)
  ada_clf = AdaBoostClassifier(
    DecisionTreeClassifier(max_depth=1), 
    algorithm="SAMME.R", random_state=42, **params)
  ada_clf.fit(X_train, y_train)
  plt.figure(figsize=(10, 5))
  plot_decision_regions(X, y, ada_clf)
  plt.show()
```
