```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.colors import ListedColormap
from mlxtend.plotting import plot_decision_regions
```

# Random Forests

```python
from sklearn.model_selection import train_test_split
from sklearn import datasets
iris = datasets.load_iris()
list(iris.keys())

X = iris["data"][:, (2, 3)]  # petal length, petal width
y = (iris["target"]).astype(np.int)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
plt.show()
```

```python
plt.hist(y_train)
plt.hist(y_test)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from mlxtend.plotting import plot_decision_regions
```

```python
m1 = RandomForestClassifier(n_estimators=500, max_leaf_nodes=16, n_jobs=-1, random_state=42)
m1.fit(X_train, y_train)
print("m1")
plt.figure(figsize=(10, 5))
plot_decision_regions(X, y, m1)
plt.show()
```

```python
m2 = RandomForestClassifier(n_estimators=50, max_leaf_nodes=2, n_jobs=-1, random_state=42)
m2.fit(X_train, y_train)
print("m2")
plt.figure(figsize=(10, 5))
plot_decision_regions(X, y, m2)
plt.show()
```

```python
m3 = RandomForestClassifier(n_estimators=5, max_leaf_nodes=2, n_jobs=-1, random_state=42)
m3.fit(X_train, y_train)
print("m3")
plt.figure(figsize=(10, 5))
plot_decision_regions(X, y, m3)
plt.show()
```
