````python
#low noise, plenty of samples, should be easy
X1, y1 = sklearn.datasets.make_moons(n_samples=1000, noise=.05)
X2, y2 = sklearn.datasets.make_moons(n_samples=1000, noise=.3)
X3, y3 = sklearn.datasets.make_moons(n_samples=200, noise=.05)
X4, y4 = sklearn.datasets.make_moons(n_samples=200, noise=.3)
````

````python
import matplotlib.pyplot as plt

plt.scatter(X1[:,0],X1[:,1],c=y1)
plt.axis('equal')
plt.show()

plt.scatter(X2[:,0],X2[:,1],c=y2)
plt.axis('equal')
plt.show()

plt.scatter(X3[:,0],X3[:,1],c=y3)
plt.axis('equal')
plt.show()

plt.scatter(X4[:,0],X4[:,1],c=y4)
plt.axis('equal')
plt.show()
````

# Zad.
Naucz perceptron na danych iris i wylicz F1 score.

````python
iris = load_iris()
X = iris.data[:, (2, 3)] 
y = (iris.target == 0).astype(np.int) 
````

````python
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
per_clf2 = Perceptron(random_state=42)
per_clf2.fit(X_train, y_train)
````

````python
plot_decision_regions(X, y, clf=per_clf2)
````

````python
from sklearn.metrics import f1_score

f1_score(y_test, per_clf2.predict(X_test))
````
